## NERC in NLTK

```python
from nltk import word_tokenize, pos_tag, ne_chunk
 
sentence = "Mark Pedersen and John Smith are working at Google " + \
           "since 1994 for $1000 per week."

x = pos_tag(word_tokenize(sentence))
res = ne_chunk(x)
print(res)
res
```

```python
res = ne_chunk(x, binary=True)
print(res)
res
```

## NERC in CoreNLP

```python
from nltk.parse.corenlp import CoreNLPParser

l = CoreNLPParser(url='http://localhost:9000').parse(word_tokenize(sentence))
ts = [t for t in l]
len(ts)
```
